```python
#T-Test 
#One sample T test
#H0=75 vs H1 not Ho
#Decision Rule : Reject Ho, if p-value < 0.05 (The test is significant) 
x=[58,56,70, 89,56, 34,98,67,89,90]
from pingouin import ttest
t1=pg.ttest(x,75)
```

```python
t1.round(2)
```

```python
t1.to_csv("t1.csv")
```

```python
#One Sided paired T test
before=[23,25,43,43,42,40]
after=[25,24,35,37,45,30]
```

```python
t2=pg.ttest(before,after, paired=True, alternative='greater')
```

```python
t2
```

```python
#Paired T test with two tails
t4=pg.ttest(before,after, paired=True)
t4
```

```python
# Independent Two Samples  T test
x=[58,56,70, 89,56, 34,98,67,89,90] # Scores of students that were given lunch
y=[70, 89,76, 45,38,78,85,80,50,90] # Scores of studends that were not given lunch
```

```python
t5=pg.ttest(x,y)
t5
```

```python
# Independent Two Samples  T test (Unequal Samples)
x1=[58,56,70, 89,56, 34,98,67,89,90] # Scores of students that were given lunch
y1=[70, 89,76, 45,38,78,85,80,50] # Scores of studends that were not given lunch
```

```python
t6=pg.ttest(x1,y1, correction=False)
t6
```

```python
# Correlation
cr=pg.corr(x,y)
cr
```

```python
cr.to_csv("Corr Result")
```

```python
## ANOVA-Analyis of Variance
#One Way Anova
df1= pg.read_dataset("anova")
df1
```

```python
# Hypothesis: H0: A=B=C vs H1: A not equal to B and C
```

```python
aov1=pg.anova(dv='Pain threshold', between='Hair color', data=df1, detailed=True)
```

```python
aov1
```

```python
#Two Way Anova 
df2=pg.read_dataset('anova2')
df2
```

```python
aov2=df2.anova(dv="Yield",between= ["Blend","Crop"])
aov2
```

```python
aov2.to_csv('AOV2')
```

```python
#Three Way ANOVA
df3=pg.read_dataset('anova3')
df3
```

```python
aov3=df3.anova(dv="Cholesterol",between= ["Sex","Risk", 'Drug'], ss_type=3)
aov3
```

```python
aov3.to_csv("aov3")
```

```python
#Two Way Mixed Model
df4=pg.read_dataset('mixed_anova')
df4
```

```python
aov4=df4.mixed_anova(dv='Scores', between="Group", subject="Subject", within='Time')
aov4
```

```python
##One Way Repeated Measure ANOVA
df5=pg.read_dataset('rm_anova_wide')
df5
```

```python
aov5= pg.rm_anova(df5)
aov5
```
